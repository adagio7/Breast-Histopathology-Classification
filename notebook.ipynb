```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import numpy as np
import glob

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
%matplotlib inline
```

#### Load Dataset

```python
data = glob.glob("./dataset/**/*.png", recursive = True)
```

```python
negative_IDC, positive_IDC = [], []

for img in data:
    # Get the last digit to find label
    if img[-5] == "0":
        negative_IDC.append(img)
    elif img[-5] == "1":
        positive_IDC.append(img)

print("Negative IDC count: ", len(negative_IDC))
print("Positive IDC count: ", len(positive_IDC))
```

```python
plt.figure(figsize = (15, 15))

for num in range(9):
    img = tf.keras.utils.load_img(positive_IDC[num], target_size=(100, 100))
    img = tf.keras.preprocessing.image.img_to_array(img)

    plt.subplot(6, 6, 2 * num + 1)
    plt.axis("off")
    plt.title("IDC (+)")
    plt.imshow(img.astype("uint8"))

for num in range(9):
    img = tf.keras.utils.load_img(negative_IDC[num], target_size=(100, 100))
    img = tf.keras.preprocessing.image.img_to_array(img)

    plt.subplot(6, 6, 2 * (num + 1))
    plt.axis("off")
    plt.title("IDC (-)")
    plt.imshow(img.astype("uint8"))
```
